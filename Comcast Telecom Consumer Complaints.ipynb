```python
#3: PROVIDE A TABLE WITH THE FREQUENCY OF COMPLAINT TYPES & #4: WHICH ARE MAX

df['Customer Complaint']=df['Customer Complaint'].str.lower()
df['Complaint Type']=0
```

```python
for i in df['Customer Complaint']:
    if "data cap" in i:
        df['Complaint Type'][df[df['Customer Complaint']==i].index]="data cap complaint"
    elif "speed" in i:
        df['Complaint Type'][df[df['Customer Complaint']==i].index]="internet speed complaint"  
    elif "bill" in i:
        df['Complaint Type'][df[df['Customer Complaint']==i].index]="billing complaint"  
    elif "service" in i:
        df['Complaint Type'][df[df['Customer Complaint']==i].index]="service complaint"  
    elif "comcast" in i:
        df['Complaint Type'][df[df['Customer Complaint']==i].index]="network complaint"  
    elif "network" in i:        
        df['Complaint Type'][df[df['Customer Complaint']==i].index]="general comcast complaint"  
    elif "internet" in i:
        df['Complaint Type'][df[df['Customer Complaint']==i].index]="internet complaint"  
    else:
        df['Complaint Type'][df[df['Customer Complaint']==i].index]=i
```

```python
top_complaints = df['Complaint Type'].value_counts().head(15).to_frame()
top_complaints['# of Tickets'] = top_complaints['Complaint Type']
top_complaints['Complaint Type'] = top_complaints.index
top_complaints['# of Tickets'].to_frame()
```

```python
#Insights:
#We can see from the above top_complaints that most of the issues coming in are about Network/Service, Billing & Data Cap
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x='Complaint Type',y='# of Tickets',data=top_complaints)
plt.xticks(rotation=90)
plt.title("Variability of Complaints")
plt.xlabel('Complaint Type')
plt.ylabel('# of Tickets')

for i in range(15):
    plt.annotate(top_complaints['# of Tickets'][i],xy=(i-0.2,top_complaints['# of Tickets'][i]+10),size=13)
plt.show()
```

```python
#7 : WHICH STATE HAS THE MAXIMUM COMPLAINTS

df['State'].value_counts()[:1]
```

```python
#8 : STATE WITH HIGHEST PERCENTAGE OF UNRESOLVED CASES

unresolved_comp=df.groupby(['State','new_status']).size().unstack().fillna(0).sort_values(by='Open',ascending=False)

unresolved_comp['unresolved_cmp_prct']=unresolved_comp['Open']/unresolved_comp['Open'].sum()*100

unresolved_comp[:1]
```

```python
#9 : PERCENTAGE OF COMPLAINTS RESOLVED TO DATE THROUGH CUSTOMER CARE CALLS AND INTERNET

resolved_comp=df.groupby(['Received Via','new_status']).size().unstack()
resolved_comp
```

```python
resolved_comp['Resolved']=resolved_comp['Closed']/(resolved_comp['Closed']+resolved_comp['Open'])*100
resolved_comp['Resolved']
```
